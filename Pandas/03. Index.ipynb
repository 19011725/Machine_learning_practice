# 3. Index
데이터에 접근할 수 있는 주소 값

```python
import pandas as pd
data = {
    '이름' : ['강민수', '배성현', '박세환', '정태경'],
    '학교' : ['세종초', '포항초', '동탄초', '덕소초'],
    '국어' : ['100', '95', '90', '85'],
    '영어' : ['85', '90', '95', '100'],
    '수학' : ['100', '95', '90', '85'],
    '과학' : ['85', '90', '95', '100'],
    '사회' : ['100', '95', '90', '85'],
    'SW특기' : ['Python', 'C', 'C++', '']
}
df = pd.DataFrame(data, index=['1번', '2번','3번','4번'])
df
```

```python
df.index
```

## Index 이름 설정

```python
df.index.name = '지원번호'
df
```

## Index 초기화

```python
df.reset_index()
```

```python
df.reset_index(drop=True) #원래 쓰던 '지원번호' 인덱스 삭제
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True) #실제 데이터에 바로 반영
```

```python
df
```

## Index 설정
지정한 column으로 Index를 설정

```python
df.set_index('이름')
```

```python
df
```

```python
df.set_index('이름', inplace=True)
df
```

## Index 정렬
Index를 기준으로 오름차순, 내림차순 정령

```python
df.sort_index() # 인덱스로 오름차순 정렬
```

```python
df.sort_index(ascending=False) # 인덱스로 내림차순 정렬
```
