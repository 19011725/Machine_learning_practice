# 14. 그룹화
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df['학교'].replace({'세종초' : '세종중', '포항초' : '세종중'}, inplace = True)
df
```

```python
df['학교'].replace({'동탄초' : '세종고', '덕소초' : '세종고', '아산초' : '세종고', '당진초' : '세종고'}, inplace = True)
df
```

```python
df.groupby('학교')
```

```python
df.groupby('학교').get_group('세종중')
```

```python
df.groupby('학교').get_group('세종고')
```

```python
df.groupby('학교').mean() #계산 가능한 데이터들의 평균 값
```

```python
df.groupby('학교').size() #각 그룹의 크기
```

```python
df.groupby('학교').size()['세종중'] 
#학교로 그룹화를 한 뒤에 능남고에 해당하는 데이터 수
```

```python
df.groupby('학교')['키'].mean() #학교로 그룹화를 한 뒤에 키의 평균 데이터
```

```python
df.groupby('학교')[['국어', '영어', '수학']].mean() 
#학교로 그룹화를 한 뒤에 국어, 영어, 수학 평균 데이터
```

```python
df['학년'] = [2,2,2,1,3,3,1] #학년 Column 추가
df
```

```python
df.groupby(['학교', '학년']).mean() #학교별, 학년별 평균 데이터
```

```python
df.groupby('학년').mean() #학년별 평균 데이터
```

```python
df.groupby('학년').mean().sort_values('키')
```

```python
df.groupby('학년').mean().sort_values('키', ascending = False)
```

```python
df.groupby(['학교', '학년']).sum()
```

```python
df.groupby('학교')[['이름', 'SW특기']].count() 
#학교로 그룹화를 한 뒤에 각 학교별 SW특기 데이터의 수를 가져옴
```

```python
school = df.groupby('학교')
school['학년'].value_counts() #학교로 그룹화를 한 뒤에 학년별 학생 수를 가져옴
```

```python
school['학년'].value_counts().loc['세종고'] 
#학교로 그룹화를 한 뒤에 세종고에 대해서 학년별 학생 수를 가져옴
```

```python
school['학년'].value_counts(normalize=True).loc['세종고']
```
