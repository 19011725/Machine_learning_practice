# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd
data = {
    '이름' : ['강민수', '배성현', '박세환', '정태경'],
    '학교' : ['세종초', '포항초', '동탄초', '덕소초'],
    '국어' : ['100', '95', '90', '85'],
    '영어' : ['85', '90', '95', '100'],
    '수학' : ['100', '95', '90', '85'],
    '과학' : ['85', '90', '95', '100'],
    '사회' : ['100', '95', '90', '85'],
    'SW특기' : ['Python', 'C', 'C++', '']
}
df = pd.DataFrame(data, index=['1번', '2번','3번','4번'])
df.index.name = '지원번호'
df
```

## 저장하기
### csv파일로 저장

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False) #지원번호 업이 저장
```

### 텍스트(.txt) 파일로 저장

```python
df.to_csv('score.txt', sep='\t') #tap으로 구분된 텍스트 파일
```

### 엑셀 파일로 저장

```python
df.to_excel('score.xlsx')
```

***
## 열기
### csv 파일 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) #1개 row를 스킵한다.
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1,3]) #1, 3 삭제 row는 0부터 시작
df
```

```python
df = pd.read_csv('score.csv', nrows = 2) #지정된 갯수 만큼의 row만 가져옴
df
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=2) #처음2개 무시, 그 밑에 2개 가져오기
df
```

### 텍스트(.txt)파일 열기

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col = '지원번호')
df
```

```python
df = pd.read_csv('score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

### 엑셀 파일 열기

```python
df = pd.read_excel('score.xlsx')
df
```

```python
df = pd.read_excel('Score.xlsx', index_col = '지원번호')
df
```
