## 데이터 제외하기 dropna

```python
df=pd.read_excel('score.xlsx', index_col='지원번호')
df.dropna() #전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

- axis : index of columns
- how : any of all

```python
df.dropna(axis='index', how='any') #NaN가 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') #NaN이 하나라도 있는 column삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') #데이터 전체가 NaN인 경우에만 Column삭제
```
