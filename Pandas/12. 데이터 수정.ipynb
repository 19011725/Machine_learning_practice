# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

## column 수정

```python
df['학교'].replace({'세종초' : '세종대', '포항초' : '포항중'}, inplace = True)
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' #학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' #결과 Column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 450, '결과'] = 'Pass' #총합이 450보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

## Column삭제

```python
df.drop(columns=['총합']) #총합 Column을 삭제
```

```python
df.drop(columns=['국어', '영어', '수학']) #국어, 영어, 수학 Column을 삭제
```

```python
df
```

## Cell 수정

```python
df.loc['4번', 'SW특기'] = 'C' #4번 학생의 SW특기 데이터를 C로 변경
df
```

```python
df.loc['5번', ['학교', 'SW특기']] = ['세종대학교', 'C#'] #번 학생의 학교, SW특기 데이터 변경
df
```
