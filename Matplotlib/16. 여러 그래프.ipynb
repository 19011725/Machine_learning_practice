# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) #세로로 n개 가로로 n개 만들겠다.
fig.suptitle('여러 그래프 넣기')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) #세로로 n개 가로로 n개 만들겠다.
fig.suptitle('여러 그래프 넣기')

#첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수')
axs[0, 0].set_title('첫 번째 그래프') #제목
axs[0, 0].legend() #범례
axs[0, 0].set(xlabel='이름', ylabel='점수') #x, y축 label
axs[0, 0].set_facecolor('lightyellow') #전경색
axs[0, 0].grid(linestyle = '--', linewidth=0.5)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) #세로로 n개 가로로 n개 만들겠다.
fig.suptitle('여러 그래프 넣기')

#첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수')
axs[0, 0].set_title('첫 번째 그래프') #제목
axs[0, 0].legend() #범례
axs[0, 0].set(xlabel='이름', ylabel='점수') #x, y축 label
axs[0, 0].set_facecolor('lightyellow') #전경색
axs[0, 0].grid(linestyle = '--', linewidth=0.5)

#두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].set_title('두 번째 그래프')
axs[0, 1].set(ylabel='점수')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) #세로로 n개 가로로 n개 만들겠다.
fig.suptitle('여러 그래프 넣기')

#첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수')
axs[0, 0].set_title('첫 번째 그래프') #제목
axs[0, 0].legend() #범례
axs[0, 0].set(xlabel='이름', ylabel='점수') #x, y축 label
axs[0, 0].set_facecolor('lightyellow') #전경색
axs[0, 0].grid(linestyle = '--', linewidth=0.5)

#두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].set_title('두 번째 그래프')
axs[0, 1].set(ylabel='점수')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

#세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'], label = '키')
axs[1, 0].set_title('세 번째 그래프')
axs[1, 0].set_xlim(165, 200)

#네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], color = 'green', alpha=0.5, label='사회')
axs[1, 1].set_title('네 번째 그래프')
axs[1, 1].legend()
```
